Crear la conexión a la base de datos Northwind que devuelva un mensaje de error si algo no ocurre cómo debería.

```python
try:
  cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1',
                              database='northwind')
# en caso de que no lo consigas por que hay algún error entonces ...
except mysql.connector.Error as err:

  # si es un error con la contraseña devuelveme un mensaje de acceso denegado ya que tenemos problemas con la contraseña
  if err.errno == errorcode.ER_ACCESS_DENIED_ERROR:
    print("Something is wrong with your user name or password")
  
  # si el error no tiene que ver con la contraseña, puede ser porque la base de datos no exista, devuelveme un mensaje de que la base de datos no existe
  elif err.errno == errorcode.ER_BAD_DB_ERROR:
    print("Database does not exist")
  
  # si no es por ninguno de los errores anteriores, printeame cual es el error que estoy teniendo en mi conexión
  else:
    print(err)
else:
  cnx.close() 
```

Usando "information_schema", extraer los nombres de las columnas de la tabla "customers" de la BBDD "northwind".

```python


cnx = mysql.connector.connect(user='root', password='AlumnaAdalab',
                              host='127.0.0.1', database='northwind')


#Creamos el servidor
mycursor = cnx.cursor()

#Ejecutamos la consulta
query= ("SELECT COLUMN_NAME FROM information_schema.columns WHERE TABLE_SCHEMA = 'northwind' AND TABLE_NAME = 'customers'")

mycursor.execute(query)  

#Recuperamos los resultados
resultados = mycursor.fetchall()  # Obtiene todos los resultados de la consulta, metodo fetchall devuelve una lista de tuplas con los nombres de las columnas

# Imprimir los resultados
for columna in resultados:
    print(columna)

# Cerramos la conexión
mycursor.close()
cnx.close()



```
